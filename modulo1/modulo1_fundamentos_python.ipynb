## 1. Variáveis e Tipos de Dados

Variáveis armazenam valores na memória do programa. Em Python, o tipo é inferido automaticamente — não é necessário declarar.

Os quatro tipos primitivos mais usados são:

| Tipo | Descrição | Exemplo |
|------|-----------|--------|
| `str` | Texto | `"Maria"` |
| `int` | Número inteiro | `22` |
| `float` | Número decimal | `1.65` |
| `bool` | Verdadeiro ou Falso | `True` |

````python
# Criando variáveis de diferentes tipos
nome = "Maria"       # str  — texto sempre entre aspas
idade = 22           # int  — número inteiro
altura = 1.65        # float — número decimal (use ponto, não vírgula)
aprovado = True      # bool — True ou False, com letra maiúscula

print("Nome:", nome)
print("Idade:", idade)
print("Altura:", altura)
print("Aprovado:", aprovado)
````

````python
# A função type() retorna o tipo de uma variável
print(type(nome))      # <class 'str'>
print(type(idade))     # <class 'int'>
print(type(altura))    # <class 'float'>
print(type(aprovado))  # <class 'bool'>
````

````python
# Operações aritméticas
a = 10
b = 3

print("Soma:", a + b)             # 13
print("Subtracao:", a - b)        # 7
print("Multiplicacao:", a * b)    # 30
print("Divisao:", a / b)          # 3.333...
print("Divisao inteira:", a // b) # 3
print("Resto:", a % b)            # 1
print("Potencia:", a ** b)        # 1000
````

````python
# f-string é a forma mais prática de montar textos com variáveis
mensagem = f"Ola, {nome}! Voce tem {idade} anos e {altura}m de altura."
print(mensagem)

# Métodos úteis de string
print(nome.upper())   # MARIA
print(nome.lower())   # maria
print(len(nome))      # 5 — quantidade de caracteres
````

## 2. Estruturas Condicionais

O `if` permite que o programa execute blocos diferentes dependendo de uma condição. A indentação (4 espaços) é obrigatória — ela define o que está dentro de cada bloco.

```python
if condicao:
    # executa se verdadeiro
elif outra_condicao:
    # executa se a primeira for falsa e esta for verdadeira
else:
    # executa se nenhuma condição anterior for verdadeira
```

````python
# Classificação de nota
nota = 7.5

if nota >= 9:
    print("Excelente")
elif nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperacao")
else:
    print("Reprovado")
````

````python
# Operadores de comparação
x = 10

print(x > 5)    # True
print(x < 5)    # False
print(x == 10)  # True  — atenção: dois sinais de = para comparar
print(x != 10)  # False — diferente de
print(x >= 10)  # True
print(x <= 10)  # True
````

````python
# Operadores lógicos: and, or, not
idade = 20
tem_carteira = True

# and — ambas as condições precisam ser verdadeiras
if idade >= 18 and tem_carteira:
    print("Pode dirigir")

# or — pelo menos uma condição precisa ser verdadeira
if idade >= 18 or tem_carteira:
    print("Pelo menos uma condicao e verdadeira")

# not — inverte o valor lógico
if not tem_carteira:
    print("Sem carteira")
else:
    print("Com carteira")
````

## 3. Estruturas de Repetição

Laços evitam repetição manual de código. Use `for` quando souber quantas iterações vai fazer, e `while` quando depender de uma condição.

````python
# for percorrendo uma lista
frutas = ["maca", "banana", "uva", "laranja"]

for fruta in frutas:
    print(fruta)
````

````python
# range() gera uma sequência de números
# range(5)    -> 0, 1, 2, 3, 4
# range(1, 6) -> 1, 2, 3, 4, 5

soma = 0
for numero in range(1, 6):
    soma += numero

print(f"Soma de 1 a 5: {soma}")
````

````python
# enumerate() retorna o índice junto com o valor
notas = [8.5, 9.0, 7.5, 6.0]

for i, nota in enumerate(notas):
    print(f"Aluno {i+1}: {nota}")
````

````python
# while repete enquanto a condição for verdadeira
# Sempre garanta que a condição vai se tornar falsa em algum momento
contador = 0

while contador < 5:
    print(f"Contador: {contador}")
    contador += 1

print("Loop encerrado")
````

````python
# break encerra o loop, continue pula para a próxima iteração
for numero in range(10):
    if numero == 3:
        continue  # pula o 3
    if numero == 7:
        break     # para no 7
    print(numero)
````

## 4. Estruturas de Dados

Python tem três estruturas principais para armazenar coleções de valores. Cada uma tem um propósito diferente.

- **Lista** — quando a ordem importa e os dados podem mudar
- **Tupla** — quando os dados não devem ser alterados
- **Dicionário** — quando você precisa acessar dados por nome (chave)

````python
# Lista [ ]
# Ordenada, mutável, permite duplicatas
# Índice começa em 0

notas = [8.5, 9.0, 7.5, 6.0, 9.0]

print(notas[0])    # primeiro elemento: 8.5
print(notas[-1])   # último elemento: 9.0
print(len(notas))  # tamanho: 5

notas.append(10.0)  # adiciona ao final
notas.remove(6.0)   # remove o valor 6.0

print("Maior:", max(notas))
print("Menor:", min(notas))
print("Media:", sum(notas) / len(notas))
````

````python
# Tupla ( )
# Ordenada, imutável

ponto = (10, 20)

print(ponto[0])  # 10
print(ponto[1])  # 20

# Desempacotamento: atribui cada valor a uma variável
x, y = ponto
print(f"x={x}, y={y}")

# Tentar modificar uma tupla gera erro:
# ponto[0] = 99  ->  TypeError
````

````python
# Dicionário { }
# Pares chave:valor, acesso por chave

aluno = {
    "nome": "Ana Silva",
    "idade": 20,
    "curso": "Engenharia",
    "notas": [8.5, 9.0, 7.5]
}

print(aluno["nome"])   # Ana Silva
print(aluno["notas"])  # [8.5, 9.0, 7.5]

aluno["email"] = "ana@email.com"  # adiciona nova chave
aluno["idade"] = 21               # atualiza valor existente

for chave, valor in aluno.items():
    print(f"{chave}: {valor}")
````

````python
# Lista de dicionários — estrutura muito comum em Ciência de Dados
turma = [
    {"nome": "Ana",   "nota": 9.0},
    {"nome": "Bruno", "nota": 7.5},
    {"nome": "Carol", "nota": 8.0},
    {"nome": "Diego", "nota": 6.5},
]

soma = 0
for estudante in turma:
    soma += estudante["nota"]

media = soma / len(turma)
print(f"Media da turma: {media:.2f}")

print("\nAprovados:")
for estudante in turma:
    if estudante["nota"] >= 7:
        print(f"  {estudante['nome']} — {estudante['nota']}")
````

## 5. Boas Práticas

Código bom não é só aquele que funciona — é aquele que outra pessoa consegue ler e entender. As convenções abaixo são padrão no mercado.

````python
# Ruim: nomes sem significado, sem comentários
x = [8.5, 9.0, 7.5, 6.0]
s = 0
for i in x:
    s += i
r = s / len(x)
print(r)
````

````python
# Bom: nomes descritivos, comentários onde fazem sentido
notas_turma = [8.5, 9.0, 7.5, 6.0]

soma_notas = 0
for nota in notas_turma:
    soma_notas += nota

media_turma = soma_notas / len(notas_turma)
print(f"Media da turma: {media_turma:.2f}")
````

````python
# Funções evitam repetição de código
# A docstring documenta o que a função faz

def calcular_media(lista_notas):
    """Retorna a media de uma lista de notas."""
    if len(lista_notas) == 0:
        return 0
    return sum(lista_notas) / len(lista_notas)

def classificar_aluno(media):
    """Retorna a situacao do aluno com base na media."""
    if media >= 9:
        return "Excelente"
    elif media >= 7:
        return "Aprovado"
    elif media >= 5:
        return "Recuperacao"
    else:
        return "Reprovado"

notas_ana = [9.0, 8.5, 9.5]
media_ana = calcular_media(notas_ana)
print(f"Ana — Media: {media_ana:.2f} — {classificar_aluno(media_ana)}")
````
