## 1. Fundamentos da Visualização de Dados

Visualizar dados não é apenas "fazer gráficos bonitos". A visualização é uma forma de **comunicação**. Ela existe para:

- **Reduzir a dificuldade de interpretação**: quando uma tabela tem milhares de linhas, o cérebro humano tem dificuldade para perceber tendências só olhando para números.
- **Destacar padrões**: gráficos permitem identificar outliers, concentrações e correlações rapidamente.
- **Apoiar a tomada de decisão**: executivos confiam mais em análises quando há suporte visual.
- **Contar histórias com dados**: toda boa visualização tem um propósito narrativo claro.

No contexto do e-commerce, por exemplo:
- Um gráfico de série temporal mostra a **sazonalidade** das vendas.
- Um mapa de calor revela **padrões de distribuição** geográfica.
- Um gráfico de dispersão pode indicar a **correlação** entre preço e avaliação.

Sem visualização, essas conclusões exigiriam cálculos estatísticos demorados.

## 2. Tipos de Gráficos: Comparação, Distribuição, Relação e Composição

Em análise de dados, quatro objetivos aparecem com frequência:

### 2.1 Comparação
A comparação é usada quando queremos verificar diferenças entre categorias ou grupos.

**Exemplos de perguntas:**
- Quais categorias de produtos têm maior receita?
- Qual estado tem mais pedidos?
- Qual é a diferença de preço entre sellers?

**Gráficos recomendados:** barras, barras horizontais, lollipop.

### 2.2 Distribuição
A distribuição mostra como os valores se espalham ao longo de um intervalo.

**Exemplos de perguntas:**
- Como os preços se distribuem?
- Qual é a forma da distribuição de avaliações?
- Há outliers nas entregas?

**Gráficos recomendados:** histogramas, densidade, box plot, violin plot.

### 2.3 Relação
A relação mostra como duas ou mais variáveis se associam.

**Exemplos de perguntas:**
- Existe correlação entre preço e avaliação?
- O custo de frete impacta o prazo de entrega?
- Como o tempo de processamento evolui com o volume?

**Gráficos recomendados:** scatter plot, linha (série temporal), heatmap de correlação.

### 2.4 Composição
A composição mostra como um total se divide em partes.

**Exemplos de perguntas:**
- Qual é a distribuição de status dos pedidos?
- Qual estado representa maior percentual de vendas?
- Como a receita se divide por categoria?

**Gráficos recomendados:** pizza, barras empilhadas, donut.

## 3. Carga Cognitiva: menos é mais na comunicação visual

**Carga cognitiva** é o esforço mental necessário para entender uma informação.

Quando um gráfico tem excesso de elementos — muitas cores, categorias, títulos vagos, poluição visual ou ausência de foco — o leitor precisa gastar mais energia para compreender a mensagem. Isso dificulta a análise e reduz o impacto da visualização.

### Dicas para reduzir carga cognitiva:

1. **Título claro e específico**: não escreva "Pedidos por Estado", escreva "Amazonas lidera em volume de pedidos (2016-2018)".

2. **Uma ideia por gráfico**: se precisa mostrar três relações, faça três gráficos.

3. **Paleta de cores reduzida**: use no máximo 3-5 cores. Cores devem ter propósito (destacar, categorizar ou sequenciar).

4. **Remova tudo que não adiciona informação**: se a grid não ajuda, remova. Se uma legenda é óbvia, remova.

5. **Rótulos diretos**: quando possível, rotule as barras ou pontos diretamente em vez de forçar o leitor a procurar na legenda.

6. **Escala apropriada**: não manipule os eixos para exagerar diferenças mínimas.

## 4. Bibliotecas práticas: Matplotlib e Seaborn

### 4.1 Matplotlib

Na prática, usaremos Seaborn para a maioria dos gráficos e recorreremos ao Matplotlib quando precisarmos de controle fino.

## 5. Importação de bibliotecas e carregamento do dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings

warnings.filterwarnings('ignore')

# Configuração visual padrão com Seaborn
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)
plt.rcParams["font.size"] = 10

print("Bibliotecas importadas com sucesso!")
print(f"Versão do Pandas: {pd.__version__}")
print(f"Versão do Matplotlib: {plt.matplotlib.__version__}")
print(f"Versão do Seaborn: {sns.__version__}")
```

### 5.1 Carregando os dados

Aqui carregaremos o dataset do Módulo 2 e realizaremos transformações básicas para preparar os dados para visualização.

```python
# Caminho para o arquivo (ajuste conforme sua estrutura de pastas)
# Se você tem os arquivos CSV do Olist, carregue-os assim:

caminhos_dados = {
    'orders': '../datasets/olist_orders_dataset.csv',
    'order_items': '../datasets/olist_order_items_dataset.csv',
    'customers': '../datasets/olist_customers_dataset.csv',
    'products': '../datasets/olist_products_dataset.csv',
    'order_reviews': '../datasets/olist_order_reviews_dataset.csv',
}

# Tentando carregar os dados
dfs = {}
for nome, caminho in caminhos_dados.items():
    if Path(caminho).exists():
        dfs[nome] = pd.read_csv(caminho)
        print(f"✓ {nome}: {dfs[nome].shape[0]} linhas")
    else:
        print(f"✗ {nome} não encontrado em {caminho}")

# Se os arquivos foram carregados, criamos um dataset consolidado
if 'orders' in dfs:
    # Exemplo de consolidação
    df_merged = dfs['orders'].copy()
    
    if 'customers' in dfs:
        df_merged = df_merged.merge(dfs['customers'][['customer_id', 'customer_state', 'customer_city']], 
                                     on='customer_id', how='left')
    
    if 'order_reviews' in dfs:
        df_merged = df_merged.merge(dfs['order_reviews'][['order_id', 'review_score']], 
                                     on='order_id', how='left')
    
    if 'order_items' in dfs:
        df_items = dfs['order_items'].groupby('order_id')['price'].sum().reset_index()
        df_items.rename(columns={'price': 'order_value'}, inplace=True)
        df_merged = df_merged.merge(df_items, on='order_id', how='left')
    
    # Conversão de datas
    df_merged['order_purchase_timestamp'] = pd.to_datetime(df_merged['order_purchase_timestamp'])
    df_merged['order_approved_at'] = pd.to_datetime(df_merged['order_approved_at'])
    df_merged['order_delivered_customer_date'] = pd.to_datetime(df_merged['order_delivered_customer_date'])
    
    # Criar coluna de mês/ano para análise temporal
    df_merged['order_year_month'] = df_merged['order_purchase_timestamp'].dt.to_period('M')
    df_merged['order_year'] = df_merged['order_purchase_timestamp'].dt.year
    df_merged['order_month'] = df_merged['order_purchase_timestamp'].dt.month
    
    print(f"\n✓ Dataset consolidado: {df_merged.shape[0]} pedidos, {df_merged.shape[1]} colunas")
    print(f"\nPrimeiras linhas:")
    print(df_merged.head())
else:
    print("\n⚠ Nenhum dataset foi carregado. Certifique-se de que os arquivos CSV estão no diretório '../datasets/'")
```

## 6. Estrutura dos dados para visualização

O dataset consolidado contém as seguintes informações principais:

### Dimensões temporais
- `order_purchase_timestamp`: quando o pedido foi feito
- `order_approved_at`: quando foi aprovado
- `order_delivered_customer_date`: quando foi entregue

### Localização
- `customer_state`: estado do cliente
- `customer_city`: cidade do cliente

### Negócio
- `order_status`: situação do pedido (delivered, cancelled, etc.)
- `order_value`: valor total do pedido
- `review_score`: avaliação do cliente (1-5)

Essas variáveis nos permitirão criar análises sobre:
1. **Série temporal**: evolução das vendas ao longo do tempo
2. **Distribuição geográfica**: padrões por estado
3. **Qualidade**: relação entre prazo, preço e satisfação
4. **Desempenho logístico**: tempo de entrega e taxa de cancelamento

## 7. Exemplos práticos de gráficos

Nos exemplos a seguir, cada bloco foi organizado para representar um tipo de objetivo visual.

A ideia é que você não apenas reproduza o código, mas também **interprete o que cada gráfico mostra** e qual pergunta de negócio ele responde.

### 7.1 COMPARAÇÃO — Gráfico de Barras

**Objetivo:** Comparar o volume de pedidos entre os principais estados.

**Por que esse gráfico?** Quando o objetivo é comparar valores entre categorias, o gráfico de barras costuma ser a melhor escolha. Ele é fácil de ler e permite ordenação natural.

**Pergunta de negócio:** Quais estados têm maior concentração de clientes?

```python
if 'df_merged' in locals():
    # Contar pedidos por estado (top 15)
    pedidos_por_estado = (
        df_merged[['customer_state']]
        .value_counts()
        .head(15)
    )
    
    # Criar figura e eixo
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Gráfico de barras horizontal (melhor para muitas categorias)
    sns.barplot(
        y=pedidos_por_estado.index,
        x=pedidos_por_estado.values,
        palette="viridis",
        ax=ax
    )
    
    # Configurações de estilo
    ax.set_title(
        "São Paulo e Minas Gerais dominam as vendas online no Brasil (2016-2018)",
        fontsize=14,
        fontweight='bold',
        pad=20
    )
    ax.set_xlabel("Número de Pedidos", fontsize=12)
    ax.set_ylabel("Estado", fontsize=12)
    
    # Adicionar valores nas barras
    for i, v in enumerate(pedidos_por_estado.values):
        ax.text(v + 500, i, str(v), va='center', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Insight: {pedidos_por_estado.index[0]} concentra {pedidos_por_estado.values[0]} pedidos,")
    print(f"   representando {(pedidos_por_estado.values[0] / df_merged.shape[0] * 100):.1f}% do total.")
```

### 7.2 COMPARAÇÃO — Gráfico de Valores de Pedidos por Status

**Objetivo:** Entender como o valor médio dos pedidos varia conforme seu status final.

**Por que esse gráfico?** Um boxplot combinado com strip plot mostra não apenas a mediana, mas também a distribuição e outliers.

**Pergunta de negócio:** Pedidos cancelados tendem a ter valores diferentes dos entregues?

```python
if 'df_merged' in locals() and 'order_value' in df_merged.columns:
    # Remover valores nulos
    df_clean = df_merged.dropna(subset=['order_value', 'order_status'])
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Boxplot combinado com pontos
    sns.boxplot(
        data=df_clean,
        x='order_status',
        y='order_value',
        palette='Set2',
        ax=ax
    )
    
    # Adicionar pontos para mostrar a distribuição real (com transparência)
    sns.stripplot(
        data=df_clean.sample(min(1000, len(df_clean))),
        x='order_status',
        y='order_value',
        color='black',
        alpha=0.2,
        size=4,
        ax=ax
    )
    
    ax.set_title(
        "Distribuição de Valores de Pedidos por Status",
        fontsize=14,
        fontweight='bold',
        pad=20
    )
    ax.set_xlabel("Status do Pedido", fontsize=12)
    ax.set_ylabel("Valor do Pedido (R$)", fontsize=12)
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    # Estatísticas por status
    print("\n📊 Estatísticas por Status:")
    print(df_clean.groupby('order_status')['order_value'].describe().round(2))
```

### 7.3 DISTRIBUIÇÃO — Histograma e Densidade de Valores

**Objetivo:** Entender como os valores de pedidos se distribuem.

**Por que esse gráfico?** Um histograma com curva de densidade mostra tanto a frequência quanto a forma da distribuição, revelando se há picos, assimetrias ou múltiplas modas.

**Pergunta de negócio:** Qual é o padrão de gasto dos clientes? Há uma concentração em faixas específicas?

```python
if 'df_merged' in locals() and 'order_value' in df_merged.columns:
    df_clean = df_merged.dropna(subset=['order_value'])
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Histograma com linha de densidade
    sns.histplot(
        data=df_clean,
        x='order_value',
        kde=True,
        stat='density',
        bins=50,
        color='skyblue',
        ax=ax
    )
    
    # Adicionar linhas verticais para média e mediana
    media = df_clean['order_value'].mean()
    mediana = df_clean['order_value'].median()
    
    ax.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Média: R${media:.2f}')
    ax.axvline(mediana, color='green', linestyle='--', linewidth=2, label=f'Mediana: R${mediana:.2f}')
    
    ax.set_title(
        "Distribuição dos Valores de Pedidos: Padrão de Gasto dos Clientes",
        fontsize=14,
        fontweight='bold',
        pad=20
    )
    ax.set_xlabel("Valor do Pedido (R$)", fontsize=12)
    ax.set_ylabel("Densidade", fontsize=12)
    ax.legend(fontsize=10)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Insights sobre distribuição:")
    print(f"  • Média: R${media:.2f}")
    print(f"  • Mediana: R${mediana:.2f}")
    print(f"  • Assimetria (skewness): {df_clean['order_value'].skew():.2f}")
    print(f"  • Curtose (kurtosis): {df_clean['order_value'].kurtosis():.2f}")
```

### 7.4 RELAÇÃO — Série Temporal de Vendas

**Objetivo:** Visualizar a evolução das vendas ao longo do tempo.

**Por que esse gráfico?** Um gráfico de linha mostra tendências, sazonalidade e quebras estruturais na série temporal.

**Pergunta de negócio:** Como evoluem as vendas? Há padrões sazonais? Períodos críticos?

```python
if 'df_merged' in locals() and 'order_year_month' in df_merged.columns:
    # Agrupar por mês
    vendas_por_mes = (
        df_merged.dropna(subset=['order_year_month', 'order_value'])
        .groupby('order_year_month')
        .agg({
            'order_id': 'count',
            'order_value': 'sum'
        })
        .rename(columns={'order_id': 'num_pedidos', 'order_value': 'receita'})
    )
    
    # Converter índice Period para string para plotagem
    vendas_por_mes.index = vendas_por_mes.index.astype(str)
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))
    
    # Gráfico 1: Número de pedidos
    ax1.plot(range(len(vendas_por_mes)), vendas_por_mes['num_pedidos'].values, 
             marker='o', linewidth=2.5, markersize=6, color='#2E86AB')
    ax1.fill_between(range(len(vendas_por_mes)), vendas_por_mes['num_pedidos'].values, 
                      alpha=0.3, color='#2E86AB')
    ax1.set_title("Volume de Pedidos ao Longo do Tempo", fontsize=12, fontweight='bold')
    ax1.set_ylabel("Número de Pedidos", fontsize=11)
    ax1.grid(True, alpha=0.3)
    ax1.set_xticks(range(0, len(vendas_por_mes), 3))
    ax1.set_xticklabels(vendas_por_mes.index[::3], rotation=45)
    
    # Gráfico 2: Receita total
    ax2.plot(range(len(vendas_por_mes)), vendas_por_mes['receita'].values, 
             marker='s', linewidth=2.5, markersize=6, color='#A23B72')
    ax2.fill_between(range(len(vendas_por_mes)), vendas_por_mes['receita'].values, 
                      alpha=0.3, color='#A23B72')
    ax2.set_title("Receita Total ao Longo do Tempo", fontsize=12, fontweight='bold')
    ax2.set_xlabel("Período", fontsize=11)
    ax2.set_ylabel("Receita (R$)", fontsize=11)
    ax2.grid(True, alpha=0.3)
    ax2.set_xticks(range(0, len(vendas_por_mes), 3))
    ax2.set_xticklabels(vendas_por_mes.index[::3], rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Padrões temporais:")
    print(f"  • Mês com mais pedidos: {vendas_por_mes['num_pedidos'].idxmax()} ({vendas_por_mes['num_pedidos'].max()} pedidos)")
    print(f"  • Mês com maior receita: {vendas_por_mes['receita'].idxmax()} (R${vendas_por_mes['receita'].max():.2f})")
```

### 7.5 RELAÇÃO — Scatter Plot: Preço vs Avaliação

**Objetivo:** Investigar se existe correlação entre o valor do pedido e a satisfação do cliente.

**Por que esse gráfico?** Um scatter plot com regressão linear permite visualizar correlações e outliers de forma clara.

**Pergunta de negócio:** Clientes que gastam mais estão mais satisfeitos? Há uma relação negativa?

```python
if 'df_merged' in locals() and 'review_score' in df_merged.columns and 'order_value' in df_merged.columns:
    df_scatter = df_merged.dropna(subset=['order_value', 'review_score'])
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Scatter plot com regressão
    sns.regplot(
        data=df_scatter,
        x='order_value',
        y='review_score',
        scatter_kws={'alpha': 0.3, 's': 20},
        line_kws={'color': 'red', 'linewidth': 2},
        ax=ax
    )
    
    ax.set_title(
        "Relação entre Valor do Pedido e Satisfação do Cliente",
        fontsize=14,
        fontweight='bold',
        pad=20
    )
    ax.set_xlabel("Valor do Pedido (R$)", fontsize=12)
    ax.set_ylabel("Avaliação do Cliente (1-5)", fontsize=12)
    
    # Calcular correlação
    correlacao = df_scatter['order_value'].corr(df_scatter['review_score'])
    ax.text(0.05, 0.95, f'Correlação de Pearson: {correlacao:.3f}', 
            transform=ax.transAxes, fontsize=11, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Análise de correlação:")
    print(f"  • Correlação de Pearson: {correlacao:.4f}")
    if abs(correlacao) < 0.3:
        print(f"  • Interpretação: Correlação fraca entre valor e avaliação")
    elif abs(correlacao) < 0.7:
        print(f"  • Interpretação: Correlação moderada")
    else:
        print(f"  • Interpretação: Correlação forte")
```

### 7.6 COMPOSIÇÃO — Pizza Chart: Status dos Pedidos

**Objetivo:** Mostrar a proporção de pedidos em cada status.

**Por que esse gráfico?** Um gráfico de pizza é útil quando há poucas categorias (até 5) e queremos destacar proporções visuais.

**Pergunta de negócio:** Qual é a taxa de sucesso? Quantos pedidos são cancelados?

```python
if 'df_merged' in locals() and 'order_status' in df_merged.columns:
    # Contar pedidos por status
    status_counts = df_merged['order_status'].value_counts()
    
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Cores customizadas
    cores = sns.color_palette("husl", len(status_counts))
    
    # Pie chart
    wedges, texts, autotexts = ax.pie(
        status_counts.values,
        labels=status_counts.index,
        autopct='%1.1f%%',
        colors=cores,
        startangle=90,
        textprops={'fontsize': 11}
    )
    
    # Deixar os textos de percentual em negrito e branco
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontweight('bold')
    
    ax.set_title(
        "Distribuição de Pedidos por Status (2016-2018)",
        fontsize=14,
        fontweight='bold',
        pad=20
    )
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Resumo por Status:")
    for status, count in status_counts.items():
        pct = (count / status_counts.sum()) * 100
        print(f"  • {status.upper()}: {count:,} pedidos ({pct:.1f}%)")
```

### 7.7 COMPOSIÇÃO — Barras Empilhadas: Composição por Estado

**Objetivo:** Mostrar como o status dos pedidos se distribui dentro de cada estado.

**Por que esse gráfico?** Uma barra empilhada permite comparar composição entre grupos mantendo visibilidade das partes.

**Pergunta de negócio:** Alguns estados têm maior taxa de cancelamento?

```python
if 'df_merged' in locals():
    # Top 10 estados
    top_estados = df_merged['customer_state'].value_counts().head(10).index
    
    # Criar tabela de cruzamento
    crosstab = pd.crosstab(
        df_merged[df_merged['customer_state'].isin(top_estados)]['customer_state'],
        df_merged[df_merged['customer_state'].isin(top_estados)]['order_status'],
        normalize='index'
    ) * 100
    
    fig, ax = plt.subplots(figsize=(14, 7))
    
    # Barra empilhada
    crosstab.plot(
        kind='barh',
        stacked=True,
        ax=ax,
        color=sns.color_palette("Set2", len(crosstab.columns))
    )
    
    ax.set_title(
        "Composição do Status dos Pedidos por Estado (%)",
        fontsize=14,
        fontweight='bold',
        pad=20
    )
    ax.set_xlabel("Percentual (%)", fontsize=12)
    ax.set_ylabel("Estado", fontsize=12)
    ax.legend(title="Status", bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Taxa de Entrega por Estado (Top 10):")
    print(crosstab.round(2))
```

### 7.8 RELAÇÃO — Heatmap de Correlação

**Objetivo:** Visualizar correlações entre variáveis numéricas.

**Por que esse gráfico?** Um heatmap mostra em uma visão integrada como diferentes métricas se associam.

**Pergunta de negócio:** Qual é a estrutura de correlação entre variáveis do negócio?

```python
if 'df_merged' in locals():
    # Selecionar colunas numéricas relevantes
    colunas_numericas = df_merged.select_dtypes(include=[np.number]).columns.tolist()
    
    if len(colunas_numericas) > 1:
        # Calcular matriz de correlação
        corr_matrix = df_merged[colunas_numericas].corr()
        
        fig, ax = plt.subplots(figsize=(10, 8))
        
        # Heatmap
        sns.heatmap(
            corr_matrix,
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            square=True,
            linewidths=1,
            cbar_kws={"shrink": 0.8},
            ax=ax
        )
        
        ax.set_title(
            "Matriz de Correlação entre Variáveis Numéricas",
            fontsize=14,
            fontweight='bold',
            pad=20
        )
        
        plt.tight_layout()
        plt.show()
```

## 8. Comunicação Visual e Ética

Visualização de dados também é uma questão de **responsabilidade**. Um gráfico pode informar corretamente, mas também pode induzir interpretações equivocadas quando é construído de forma ruim.

### 8.1 Como evitar gráficos enganosos

#### ❌ NÃO faça:
- **Manipular escalas** para exagerar diferenças mínimas
- **Omitir o zero** em gráficos de barras (exceto em heatmaps)
- **Usar cores** sem propósito (cores devem categorizar ou destacar)
- **Truncar eixos** para fazer um padrão parecer mais forte
- **Omitir contexto relevante** (por exemplo, esconder valores negativos)
- **Usar gráfico de pizza com mais de 5 categorias** (fica ilegível)
- **Deixar de documentar filtros ou agregações**

#### ✅ FAÇA:
- Sempre começar os eixos em zero (ou deixar bem claro por que não)
- Incluir título descritivo e fonte dos dados
- Usar cores consistentemente
- Fornecer rótulos claros
- Indicar quando há dados filtrados ou faltando
- Escolher o tipo de gráfico apropriado para a pergunta
- Testar a visualização com usuários não técnicos

### 8.2 Exemplo: o perigo de manipular escalas

```python
# Exemplo: gráfico manipulado vs. honesto
dados_exemplo = pd.DataFrame({
    'Trimestre': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Vendas': [100, 102, 105, 103]
})

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico ENGANOSO (escala manipulada)
ax1.bar(dados_exemplo['Trimestre'], dados_exemplo['Vendas'], color='red', alpha=0.7)
ax1.set_ylim(99, 106)  # ← Truncando a escala!
ax1.set_title("❌ ENGANOSO: Escala manipulada\n(parece crescimento explosivo)", 
              fontsize=12, fontweight='bold', color='red')
ax1.set_ylabel("Vendas")

# Gráfico HONESTO (escala apropriada)
ax2.bar(dados_exemplo['Trimestre'], dados_exemplo['Vendas'], color='green', alpha=0.7)
ax2.set_ylim(0, 110)  # ← Começando do zero
ax2.set_title("✅ HONESTO: Escala apropriada\n(crescimento modesto visível)", 
              fontsize=12, fontweight='bold', color='green')
ax2.set_ylabel("Vendas")

plt.suptitle("O perigo de manipular escalas em visualizações", 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("\n⚠️  Lição: A mesma informação pode ser apresentada de formas radicalmente diferentes.")
print("   Como visualizador, você tem a responsabilidade de ser honesto com os dados.")
```

## 9. Checklist: Como fazer uma visualização de qualidade

Antes de apresentar um gráfico, pergunte a si mesmo:

### Clareza e Propósito
- [ ] O gráfico responde a uma pergunta clara?
- [ ] O título é específico e informativo (não apenas "Vendas por Estado")?
- [ ] O público-alvo consegue entender sem ajuda adicional?

### Escolha do Gráfico
- [ ] É o tipo de gráfico mais apropriado para essa pergunta?
- [ ] Estou comparando, mostrando distribuição, relação ou composição?
- [ ] Evitei gráficos "bonitos mas inúteis"?

### Design Visual
- [ ] As cores têm propósito (não são apenas decoração)?
- [ ] O gráfico funciona em preto e branco (acessibilidade)?
- [ ] Os rótulos são legíveis (tamanho, rotação, sobreposição)?
- [ ] A paleta de cores respeita usuários com daltonismo?

### Dados e Honestidade
- [ ] A escala começa em zero (quando apropriado)?
- [ ] Não há manipulação intencional para exagerar efeitos?
- [ ] A fonte dos dados está clara?
- [ ] Filtros ou agregações estão documentados?
- [ ] Valores faltantes ou nulos foram tratados transparentemente?

### Contexto
- [ ] O gráfico está inserido em uma narrativa maior?
- [ ] Há anotações para destacar pontos importantes?
- [ ] A legenda é necessária ou redundante?
