## 1. Fundamentos da Visualização de Dados

Visualizar dados não é apenas "fazer gráficos bonitos". A visualização é uma forma de **comunicação**. Ela existe para:

- **Reduzir a dificuldade de interpretação**: quando uma tabela tem milhares de linhas, o cérebro humano tem dificuldade para perceber tendências só olhando para números.
- **Destacar padrões**: gráficos permitem identificar outliers, concentrações e correlações rapidamente.
- **Apoiar a tomada de decisão**: executivos confiam mais em análises quando há suporte visual.
- **Contar histórias com dados**: toda boa visualização tem um propósito narrativo claro.

No contexto do e-commerce, por exemplo:
- Um gráfico de série temporal mostra a **sazonalidade** das vendas.
- Um mapa de calor revela **padrões de distribuição** geográfica.
- Um gráfico de dispersão pode indicar a **correlação** entre preço e avaliação.

Sem visualização, essas conclusões exigiriam cálculos estatísticos demorados.

## 2. Tipos de Gráficos: Comparação, Distribuição, Relação e Composição

Em análise de dados, quatro objetivos aparecem com frequência:

### 2.1 Comparação
A comparação é usada quando queremos verificar diferenças entre categorias ou grupos.

**Exemplos de perguntas:**
- Quais categorias de produtos têm maior receita?
- Qual estado tem mais pedidos?

**Gráficos recomendados:** barras, barras horizontais, lollipop.

### 2.2 Distribuição
A distribuição mostra como os valores se espalham ao longo de um intervalo.

**Exemplos de perguntas:**
- Como os preços se distribuem?
- Qual é a forma da distribuição de avaliações?

**Gráficos recomendados:** histogramas, densidade, box plot, violin plot.

### 2.3 Relação
A relação mostra como duas ou mais variáveis se associam.

**Exemplos de perguntas:**
- Existe correlação entre preço e avaliação?
- O custo de frete impacta o prazo de entrega?

**Gráficos recomendados:** scatter plot, linha (série temporal), heatmap, regressão linear.

### 2.4 Composição
A composição mostra como um total se divide em partes.

**Exemplos de perguntas:**
- Qual é a distribuição de status dos pedidos?
- Como a receita se divide por categoria?

**Gráficos recomendados:** pizza, barras empilhadas, donut.

## 3. Carga Cognitiva: menos é mais na comunicação visual

**Carga cognitiva** é o esforço mental necessário para entender uma informação.

Quando um gráfico tem excesso de elementos, o leitor precisa gastar mais energia para compreender a mensagem.

### Dicas para reduzir carga cognitiva:

1. **Título claro e específico**: não escreva "Pedidos por Estado", escreva "São Paulo concentra 42% dos pedidos da Olist".
2. **Uma ideia por gráfico**: se precisa mostrar três relações, faça três gráficos.
3. **Paleta de cores reduzida**: use no máximo 3-5 cores.
4. **Remova tudo que não adiciona informação**.
5. **Rótulos diretos**: rotule as barras diretamente em vez de forçar o leitor a procurar na legenda.
6. **Escala apropriada**: não manipule os eixos para exagerar diferenças mínimas.

## 4. Bibliotecas práticas: Matplotlib e Seaborn

### 4.1 Matplotlib
O Matplotlib é a biblioteca base de visualização em Python. Com ela, é possível criar gráficos do zero e controlar elementos como tamanho da figura, eixos, rótulos, títulos, legendas e grids.

Aprender Matplotlib é importante porque ajuda a entender a **estrutura fundamental** de um gráfico.

### 4.2 Seaborn
O Seaborn foi construído **sobre** Matplotlib e oferece uma interface mais simples e elegante para criar gráficos estatísticos. Tem temas pré-configurados, integração nativa com Pandas e funções para gráficos complexos (heatmaps, violin plots, regressão).

## 5. Anatomia de um Gráfico Matplotlib

Todo gráfico tem uma estrutura hierárquica:

```
Figure (a "tela" completa)
 └── Axes (a área de plotagem)
      ├── Title (título)
      ├── X-axis / Y-axis (eixos)
      ├── Legend (legenda)
      ├── Grid (linhas de grade)
      └── Spines (bordas)
```

- **`Figure`**: é a "tela" onde tudo é desenhado.
- **`Axes`**: é uma área de plotagem dentro da Figure.
- **`Axis`**: refere-se aos eixos X ou Y individualmente.

Quando você escreve `fig, ax = plt.subplots()`, você está criando uma Figure e um Axes.

````python
import matplotlib.pyplot as plt

# Dados de exemplo
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [120, 140, 155, 160, 175, 200]

# 1) CRIAR FIGURE E AXES
fig, ax = plt.subplots(figsize=(10, 5))

# 2) PLOTAR DADOS
ax.plot(meses, vendas, marker='o', linewidth=2, color='#2E86AB', label='Vendas 2024')

# 3) CUSTOMIZAR
ax.set_title('Evolução das Vendas no 1º Semestre', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Mês', fontsize=12)
ax.set_ylabel('Vendas (R$ mil)', fontsize=12)
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)
ax.set_ylim(0, 250)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
````

## 6. Design de Informação: Escolha de Cores

Existem **três famílias principais** de paletas:

### 🎨 Paletas Qualitativas (categóricas)
Para categorias distintas sem ordem natural (ex: status de pedido, categoria de produto).
**Recomendadas:** `Set1`, `Set2`, `Paired`, `tab10`

### 🌈 Paletas Sequenciais
Para valores numéricos ordenados (ex: volume de vendas, densidade).
**Recomendadas:** `Blues`, `Greens`, `viridis`, `plasma`

### ↔️ Paletas Divergentes
Para valores com ponto central (positivo vs negativo).
**Recomendadas:** `RdBu`, `coolwarm`, `RdYlGn`

### Regras de ouro
1. Não use mais que 7 cores numa única visualização
2. Considere daltonismo
3. Seja consistente
4. Use intensidade para hierarquizar

````python
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

# 1) QUALITATIVA
categorias = ['Moda', 'Eletrônicos', 'Casa', 'Beleza', 'Livros']
valores_q = [45, 78, 52, 31, 24]
axes[0].bar(categorias, valores_q, color=sns.color_palette('Set2', 5))
axes[0].set_title('Qualitativa (Set2)', fontweight='bold')
axes[0].tick_params(axis='x', rotation=45)

# 2) SEQUENCIAL
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
valores_s = [20, 35, 48, 62, 75, 90]
axes[1].bar(meses, valores_s, color=sns.color_palette('Blues', 6))
axes[1].set_title('Sequencial (Blues)', fontweight='bold')

# 3) DIVERGENTE
regioes = ['NE', 'N', 'CO', 'SE', 'S']
variacao = [-15, -8, 2, 12, 18]
cores_d = sns.color_palette('RdBu', 5)
axes[2].bar(regioes, variacao, color=cores_d)
axes[2].axhline(0, color='black', linewidth=0.8)
axes[2].set_title('Divergente (RdBu)', fontweight='bold')

plt.tight_layout()
plt.show()
````

## 7. Importação e carregamento do dataset

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Caminhos para os arquivos CSV
BASE = Path("../datasets")
orders = pd.read_csv(BASE / "olist_orders_dataset.csv")
customers = pd.read_csv(BASE / "olist_customers_dataset.csv")
reviews = pd.read_csv(BASE / "olist_order_reviews_dataset.csv")
items = pd.read_csv(BASE / "olist_order_items_dataset.csv")

# Converter datas
orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])
orders["order_delivered_customer_date"] = pd.to_datetime(orders["order_delivered_customer_date"], errors="coerce")

# Consolidar em um DataFrame
df = orders.merge(customers[["customer_id", "customer_state"]], on="customer_id", how="left")
df = df.merge(reviews[["order_id", "review_score"]], on="order_id", how="left")
valor = items.groupby("order_id")["price"].sum().reset_index().rename(columns={"price": "valor"})
df = df.merge(valor, on="order_id", how="left")

# Features adicionais
df["ano_mes"] = df["order_purchase_timestamp"].dt.to_period("M").astype(str)
df["prazo_entrega"] = (df["order_delivered_customer_date"] - df["order_purchase_timestamp"]).dt.days

print("Dataset carregado:", df.shape)
df.head()
````

## 8. Exemplos práticos de gráficos

Nos exemplos a seguir, cada bloco representa um tipo de objetivo visual.

### 8.1 Comparação — gráfico de barras

**Objetivo:** comparar volume de pedidos entre os principais estados.

````python
top_estados = df["customer_state"].value_counts().head(10)

fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(y=top_estados.index, x=top_estados.values, palette="viridis", ax=ax)

ax.set_title("Top 10 Estados por Número de Pedidos", fontsize=13, fontweight="bold")
ax.set_xlabel("Número de pedidos")
ax.set_ylabel("Estado")

for i, v in enumerate(top_estados.values):
    ax.text(v + 300, i, f"{v:,}", va="center", fontweight="bold")

plt.tight_layout()
plt.show()
````

### 8.2 Distribuição — histograma

**Objetivo:** entender como os valores dos pedidos se distribuem.

````python
valores = df["valor"].dropna()
valores_lim = valores[valores <= valores.quantile(0.99)]  # limitar outliers

fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(valores_lim, bins=50, kde=True, color="skyblue", ax=ax)

ax.axvline(valores_lim.mean(), color="red", linestyle="--", linewidth=2,
           label=f"Média: R${valores_lim.mean():.2f}")
ax.axvline(valores_lim.median(), color="green", linestyle="--", linewidth=2,
           label=f"Mediana: R${valores_lim.median():.2f}")

ax.set_title("Distribuição dos Valores de Pedidos", fontsize=13, fontweight="bold")
ax.set_xlabel("Valor do pedido (R$)")
ax.set_ylabel("Frequência")
ax.legend()

plt.tight_layout()
plt.show()
````

### 8.3 Distribuição por grupos — boxplot

**Objetivo:** comparar a distribuição dos valores de pedidos entre os principais status.

````python
top_status = df["order_status"].value_counts().head(5).index
df_box = df[df["order_status"].isin(top_status)].dropna(subset=["valor"])
df_box = df_box[df_box["valor"] <= df_box["valor"].quantile(0.99)]

fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(data=df_box, x="order_status", y="valor", palette="Set2", ax=ax)

ax.set_title("Distribuição de Valores por Status do Pedido", fontsize=13, fontweight="bold")
ax.set_xlabel("Status")
ax.set_ylabel("Valor (R$)")

plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
````

### 8.4 Distribuição detalhada — violin plot

**Objetivo:** visualizar a **forma** da distribuição das avaliações por faixa de prazo.

````python
def faixa(d):
    if pd.isna(d): return np.nan
    if d <= 7:  return "1-Rápido (≤7d)"
    if d <= 15: return "2-Normal (8-15d)"
    if d <= 30: return "3-Lento (16-30d)"
    return "4-Muito lento (>30d)"

df["faixa_prazo"] = df["prazo_entrega"].apply(faixa)
df_vio = df.dropna(subset=["review_score", "faixa_prazo"])
ordem = ["1-Rápido (≤7d)", "2-Normal (8-15d)", "3-Lento (16-30d)", "4-Muito lento (>30d)"]

fig, ax = plt.subplots(figsize=(12, 6))
sns.violinplot(data=df_vio, x="faixa_prazo", y="review_score",
               order=ordem, palette="RdYlGn_r", ax=ax)

ax.set_title("Distribuição das Notas por Faixa de Prazo de Entrega",
             fontsize=13, fontweight="bold")
ax.set_xlabel("Faixa de prazo")
ax.set_ylabel("Nota de avaliação (1-5)")

plt.tight_layout()
plt.show()
````

### 8.5 Relação — scatterplot com regressão

**Objetivo:** investigar a relação entre prazo de entrega e avaliação.

````python
df_rel = df.dropna(subset=["prazo_entrega", "review_score"])
df_rel = df_rel[(df_rel["prazo_entrega"] >= 0) & (df_rel["prazo_entrega"] <= 60)]
amostra = df_rel.sample(min(5000, len(df_rel)), random_state=42)

fig, ax = plt.subplots(figsize=(12, 6))
sns.regplot(data=amostra, x="prazo_entrega", y="review_score",
            scatter_kws={"alpha": 0.2, "s": 15, "color": "#457B9D"},
            line_kws={"color": "#E63946", "linewidth": 2.5},
            ax=ax)

corr = df_rel["prazo_entrega"].corr(df_rel["review_score"])
ax.set_title(f"Prazo × Avaliação (correlação = {corr:.2f})",
             fontsize=13, fontweight="bold")
ax.set_xlabel("Prazo de entrega (dias)")
ax.set_ylabel("Nota")

plt.tight_layout()
plt.show()
````

### 8.6 Relação temporal — série temporal

**Objetivo:** visualizar a evolução das vendas ao longo do tempo.

````python
pedidos_por_mes = df.groupby("ano_mes").size().reset_index(name="num_pedidos")

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(range(len(pedidos_por_mes)), pedidos_por_mes["num_pedidos"],
        marker="o", linewidth=2.5, color="#2E86AB")
ax.fill_between(range(len(pedidos_por_mes)), pedidos_por_mes["num_pedidos"],
                alpha=0.3, color="#2E86AB")

ax.set_title("Evolução do Volume de Pedidos", fontsize=13, fontweight="bold")
ax.set_xlabel("Período")
ax.set_ylabel("Número de pedidos")
ax.set_xticks(range(0, len(pedidos_por_mes), 2))
ax.set_xticklabels(pedidos_por_mes["ano_mes"].iloc[::2], rotation=45)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

### 8.7 Relação entre múltiplas variáveis — heatmap de correlação

**Objetivo:** mostrar como todas as variáveis numéricas se correlacionam entre si.

````python
cols = ["valor", "prazo_entrega", "review_score"]
df_corr = df[cols].dropna()
corr_matrix = df_corr.corr()

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            center=0, square=True, linewidths=1, vmin=-1, vmax=1, ax=ax)

ax.set_title("Matriz de Correlação entre Variáveis", fontsize=13, fontweight="bold")

plt.tight_layout()
plt.show()
````

### 8.8 Composição — gráfico de pizza

**Objetivo:** mostrar a proporção de pedidos por status.

````python
status_counts = df["order_status"].value_counts()

fig, ax = plt.subplots(figsize=(10, 8))
cores = sns.color_palette("husl", len(status_counts))

wedges, texts, autotexts = ax.pie(
    status_counts.values, labels=status_counts.index,
    autopct="%1.1f%%", colors=cores, startangle=90
)

for autotext in autotexts:
    autotext.set_color("white")
    autotext.set_fontweight("bold")

ax.set_title("Distribuição de Pedidos por Status", fontsize=13, fontweight="bold")

plt.tight_layout()
plt.show()
````

### 8.9 Composição comparativa — barras empilhadas por estado

**Objetivo:** mostrar como o status dos pedidos se distribui em cada estado.

````python
top_estados_lista = df["customer_state"].value_counts().head(10).index
df_emp = df[df["customer_state"].isin(top_estados_lista)]

crosstab = pd.crosstab(df_emp["customer_state"], df_emp["order_status"],
                       normalize="index") * 100

fig, ax = plt.subplots(figsize=(14, 7))
crosstab.plot(kind="barh", stacked=True, ax=ax,
              color=sns.color_palette("Set2", len(crosstab.columns)))

ax.set_title("Composição do Status dos Pedidos por Estado (%)",
             fontsize=13, fontweight="bold")
ax.set_xlabel("Percentual (%)")
ax.set_ylabel("Estado")
ax.legend(title="Status", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()
plt.show()
````

## 9. Comunicação Visual e Ética

Visualização de dados também é uma questão de **responsabilidade**. Um gráfico pode informar corretamente, mas também induzir interpretações equivocadas.

### ❌ NÃO faça:
- Manipular escalas para exagerar diferenças mínimas
- Omitir o zero em gráficos de barras
- Usar cores sem propósito
- Usar gráfico de pizza com mais de 5 categorias

### ✅ FAÇA:
- Começar os eixos em zero (quando apropriado)
- Incluir título descritivo e fonte dos dados
- Usar cores consistentemente
- Escolher o tipo de gráfico apropriado

````python
# Exemplo: gráfico manipulado vs. honesto
dados_exemplo = pd.DataFrame({
    "Trimestre": ["Q1", "Q2", "Q3", "Q4"],
    "Vendas": [100, 102, 105, 103]
})

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# ENGANOSO
ax1.bar(dados_exemplo["Trimestre"], dados_exemplo["Vendas"], color="red", alpha=0.7)
ax1.set_ylim(99, 106)
ax1.set_title("❌ ENGANOSO: escala manipulada", fontweight="bold", color="red")
ax1.set_ylabel("Vendas")

# HONESTO
ax2.bar(dados_exemplo["Trimestre"], dados_exemplo["Vendas"], color="green", alpha=0.7)
ax2.set_ylim(0, 110)
ax2.set_title("✅ HONESTO: escala apropriada", fontweight="bold", color="green")
ax2.set_ylabel("Vendas")

plt.tight_layout()
plt.show()
````
