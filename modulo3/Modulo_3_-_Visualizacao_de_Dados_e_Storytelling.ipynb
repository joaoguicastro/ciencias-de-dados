## 1. Fundamentos da Visualização de Dados

Visualizar dados não é apenas "fazer gráficos bonitos". A visualização é uma forma de **comunicação**. Ela existe para:

- **Reduzir a dificuldade de interpretação**: quando uma tabela tem milhares de linhas, o cérebro humano tem dificuldade para perceber tendências só olhando para números.
- **Destacar padrões**: gráficos permitem identificar outliers, concentrações e correlações rapidamente.
- **Apoiar a tomada de decisão**: executivos confiam mais em análises quando há suporte visual.
- **Contar histórias com dados**: toda boa visualização tem um propósito narrativo claro.

No contexto do e-commerce, por exemplo:
- Um gráfico de série temporal mostra a **sazonalidade** das vendas.
- Um mapa de calor revela **padrões de distribuição** geográfica.
- Um gráfico de dispersão pode indicar a **correlação** entre preço e avaliação.

Sem visualização, essas conclusões exigiriam cálculos estatísticos demorados.

## 2. Tipos de Gráficos: Comparação, Distribuição, Relação e Composição

Em análise de dados, quatro objetivos aparecem com frequência:

### 2.1 Comparação
A comparação é usada quando queremos verificar diferenças entre categorias ou grupos.

**Exemplos de perguntas:**
- Quais categorias de produtos têm maior receita?
- Qual estado tem mais pedidos?
- Qual é a diferença de preço entre sellers?

**Gráficos recomendados:** barras, barras horizontais, lollipop.

### 2.2 Distribuição
A distribuição mostra como os valores se espalham ao longo de um intervalo.

**Exemplos de perguntas:**
- Como os preços se distribuem?
- Qual é a forma da distribuição de avaliações?
- Há outliers nas entregas?

**Gráficos recomendados:** histogramas, densidade, box plot, violin plot.

### 2.3 Relação
A relação mostra como duas ou mais variáveis se associam.

**Exemplos de perguntas:**
- Existe correlação entre preço e avaliação?
- O custo de frete impacta o prazo de entrega?
- Como o tempo de processamento evolui com o volume?

**Gráficos recomendados:** scatter plot, linha (série temporal), heatmap de correlação, regressão linear.

### 2.4 Composição
A composição mostra como um total se divide em partes.

**Exemplos de perguntas:**
- Qual é a distribuição de status dos pedidos?
- Qual estado representa maior percentual de vendas?
- Como a receita se divide por categoria?

**Gráficos recomendados:** pizza, barras empilhadas, donut.

## 3. Carga Cognitiva: menos é mais na comunicação visual

**Carga cognitiva** é o esforço mental necessário para entender uma informação.

Quando um gráfico tem excesso de elementos — muitas cores, categorias, títulos vagos, poluição visual ou ausência de foco — o leitor precisa gastar mais energia para compreender a mensagem. Isso dificulta a análise e reduz o impacto da visualização.

### Dicas para reduzir carga cognitiva:

1. **Título claro e específico**: não escreva "Pedidos por Estado", escreva "São Paulo concentra 42% dos pedidos da Olist (2016-2018)".

2. **Uma ideia por gráfico**: se precisa mostrar três relações, faça três gráficos.

3. **Paleta de cores reduzida**: use no máximo 3-5 cores. Cores devem ter propósito (destacar, categorizar ou sequenciar).

4. **Remova tudo que não adiciona informação**: se a grid não ajuda, remova. Se uma legenda é óbvia, remova.

5. **Rótulos diretos**: quando possível, rotule as barras ou pontos diretamente em vez de forçar o leitor a procurar na legenda.

6. **Escala apropriada**: não manipule os eixos para exagerar diferenças mínimas.

## 5. Anatomia de um Gráfico Matplotlib

Antes de fazer gráficos bonitos, precisamos entender **como um gráfico é construído por dentro**. No Matplotlib, todo gráfico tem uma estrutura hierárquica:

```
Figure (a "tela" completa)
 └── Axes (a área de plotagem — pode ter vários numa Figure)
      ├── Title (título)
      ├── X-axis (eixo horizontal)
      │    ├── xlabel (rótulo do eixo)
      │    └── xticks (marcações)
      ├── Y-axis (eixo vertical)
      │    ├── ylabel
      │    └── yticks
      ├── Legend (legenda)
      ├── Grid (linhas de grade)
      └── Spines (bordas)
```

### Conceitos-chave:
- **`Figure`**: é a "tela" onde tudo é desenhado. Pense nela como a folha de papel.
- **`Axes`**: é uma área de plotagem dentro da Figure. Uma Figure pode ter várias Axes (subplots).
- **`Axis`**: refere-se aos eixos X ou Y individualmente.

### Por que isso importa?
Quando você escreve `fig, ax = plt.subplots()`, você está criando uma Figure e um Axes. Todas as customizações (título, rótulos, legendas) são aplicadas ao Axes. Entender essa hierarquia é o que diferencia alguém que "copia e cola" código de alguém que **realmente sabe fazer gráficos**.

### 5.1 Demonstração prática da anatomia

Vamos criar um gráfico simples e identificar **cada elemento** para fixar o conceito.

````python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [120, 140, 155, 160, 175, 200]

# 1) CRIAR FIGURE E AXES
fig, ax = plt.subplots(figsize=(10, 5))

# 2) PLOTAR DADOS NO AXES
ax.plot(meses, vendas, marker='o', linewidth=2, color='#2E86AB', label='Vendas 2024')

# 3) CUSTOMIZAR O AXES
ax.set_title('Evolução das Vendas no 1º Semestre', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Mês', fontsize=12)
ax.set_ylabel('Vendas (R$ mil)', fontsize=12)

# 4) LEGENDA
ax.legend(loc='upper left', fontsize=11)

# 5) GRID (linhas de grade)
ax.grid(True, alpha=0.3, linestyle='--')

# 6) LIMITES DOS EIXOS
ax.set_ylim(0, 250)

# 7) REMOVER BORDAS SUPERIOR E DIREITA (boa prática de design)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 8) ANOTAÇÃO em um ponto específico
ax.annotate('Maior venda!', xy=('Jun', 200), xytext=('Abr', 220),
            arrowprops=dict(arrowstyle='->', color='red'),
            fontsize=10, color='red')

plt.tight_layout()
plt.show()

print("✓ Cada elemento deste gráfico foi criado manualmente — entender isso é essencial.")
````

## 6. Design de Informação: Escolha de Cores

A cor é um dos recursos visuais mais poderosos — e também mais mal utilizados. Usar a paleta errada pode **distorcer a mensagem** do gráfico.

### 6.1 Tipos de paletas de cores

Existem **três famílias principais** de paletas, e cada uma tem um uso específico:

#### 🎨 Paletas Qualitativas (categóricas)
Usadas para representar **categorias distintas** sem ordem natural.
- **Exemplos:** status de pedido, categoria de produto, estado do cliente
- **Paletas recomendadas:** `Set1`, `Set2`, `Paired`, `tab10`

#### 🌈 Paletas Sequenciais
Usadas para representar **valores numéricos ordenados** (de baixo para alto).
- **Exemplos:** volume de vendas, densidade populacional, temperatura
- **Paletas recomendadas:** `Blues`, `Greens`, `viridis`, `plasma`

#### ↔️ Paletas Divergentes
Usadas para representar valores com um **ponto central de referência** (positivo vs negativo, acima vs abaixo da média).
- **Exemplos:** variação de lucro, correlações, desvios em torno de zero
- **Paletas recomendadas:** `RdBu`, `coolwarm`, `RdYlGn`

### 6.2 Regras de ouro para cores
1. **Não use mais que 7 cores** numa única visualização
2. **Considere daltonismo** — teste suas cores em simuladores
3. **Seja consistente** — se azul = "bom" num gráfico, mantenha essa convenção
4. **Use intensidade para hierarquizar** — cor forte para o destaque, tom neutro para o resto

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

# 1) PALETA QUALITATIVA — categorias sem ordem
categorias = ['Moda', 'Eletrônicos', 'Casa', 'Beleza', 'Livros']
valores_q = [45, 78, 52, 31, 24]
cores_q = sns.color_palette('Set2', n_colors=len(categorias))
axes[0].bar(categorias, valores_q, color=cores_q)
axes[0].set_title('Qualitativa (Set2)\nCategorias sem ordem', fontweight='bold')
axes[0].tick_params(axis='x', rotation=45)

# 2) PALETA SEQUENCIAL — valores ordenados
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
valores_s = [20, 35, 48, 62, 75, 90]
cores_s = sns.color_palette('Blues', n_colors=len(meses))
axes[1].bar(meses, valores_s, color=cores_s)
axes[1].set_title('Sequencial (Blues)\nIntensidade = magnitude', fontweight='bold')

# 3) PALETA DIVERGENTE — positivos vs negativos
regioes = ['NE', 'N', 'CO', 'SE', 'S']
variacao = [-15, -8, 2, 12, 18]  # variação em torno de zero
cores_d = sns.color_palette('RdBu', n_colors=len(regioes))
# Ordenar cores para refletir valor
cores_ord = [cores_d[int((v - min(variacao)) / (max(variacao) - min(variacao)) * (len(cores_d)-1))] for v in variacao]
axes[2].bar(regioes, variacao, color=cores_ord)
axes[2].axhline(0, color='black', linewidth=0.8)
axes[2].set_title('Divergente (RdBu)\nPositivos vs Negativos', fontweight='bold')

plt.tight_layout()
plt.show()

print("✓ Cada tipo de paleta serve para um propósito específico.")
print("✓ Qualitativa = categorias | Sequencial = ordem | Divergente = pontos em torno de um centro")
````

## 7. Importação de bibliotecas e carregamento do dataset

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings

warnings.filterwarnings('ignore')

# Configuração visual padrão com Seaborn
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)
plt.rcParams["font.size"] = 10

print("Bibliotecas importadas com sucesso!")
print(f"Versão do Pandas: {pd.__version__}")
print(f"Versão do Matplotlib: {plt.matplotlib.__version__}")
print(f"Versão do Seaborn: {sns.__version__}")
````

### 7.1 Carregando os dados

Aqui carregaremos o dataset do Módulo 2 e realizaremos transformações básicas para preparar os dados para visualização.

````python
# Caminhos para os arquivos CSV (ajuste conforme sua estrutura)
caminhos_dados = {
    'orders': '../datasets/olist_orders_dataset.csv',
    'order_items': '../datasets/olist_order_items_dataset.csv',
    'customers': '../datasets/olist_customers_dataset.csv',
    'products': '../datasets/olist_products_dataset.csv',
    'order_reviews': '../datasets/olist_order_reviews_dataset.csv',
}

# Carregamento das tabelas
dfs = {}
for nome, caminho in caminhos_dados.items():
    if Path(caminho).exists():
        dfs[nome] = pd.read_csv(caminho)
        print(f"✓ {nome}: {dfs[nome].shape[0]:,} linhas")
    else:
        print(f"✗ {nome} não encontrado em {caminho}")

# Consolidação em um único DataFrame
df = None
if 'orders' in dfs:
    df = dfs['orders'].copy()
    
    # Merge com customers (localização)
    if 'customers' in dfs:
        df = df.merge(
            dfs['customers'][['customer_id', 'customer_state', 'customer_city']],
            on='customer_id', how='left'
        )
    
    # Merge com reviews (avaliações)
    if 'order_reviews' in dfs:
        df = df.merge(
            dfs['order_reviews'][['order_id', 'review_score']],
            on='order_id', how='left'
        )
    
    # Agregar itens do pedido (valor total)
    if 'order_items' in dfs:
        itens_agg = dfs['order_items'].groupby('order_id').agg(
            order_value=('price', 'sum'),
            num_items=('product_id', 'count')
        ).reset_index()
        df = df.merge(itens_agg, on='order_id', how='left')
    
    # Engenharia de features temporais
    df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
    df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'], errors='coerce')
    df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'], errors='coerce')
    
    df['order_year_month'] = df['order_purchase_timestamp'].dt.to_period('M')
    df['order_year'] = df['order_purchase_timestamp'].dt.year
    df['order_month'] = df['order_purchase_timestamp'].dt.month
    
    # Prazo de entrega em dias
    df['prazo_entrega_dias'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days
    
    print(f"\n✓ Dataset consolidado: {df.shape[0]:,} pedidos, {df.shape[1]} colunas")
    print(f"\nPrimeiras linhas:")
    print(df.head())
else:
    print("\n⚠ Nenhum dataset carregado. Verifique os caminhos em '../datasets/'")
````

## 9. Exemplos práticos de gráficos

Nos exemplos a seguir, cada bloco foi organizado para representar um tipo de objetivo visual.

### 9.1 COMPARAÇÃO — Gráfico de Barras

**Objetivo:** Comparar o volume de pedidos entre os principais estados.

**Por que esse gráfico?** Quando o objetivo é comparar valores entre categorias, o gráfico de barras costuma ser a melhor escolha. Ele é fácil de ler e permite ordenação natural.

**Pergunta de negócio:** Quais estados têm maior concentração de clientes?

````python
if df is not None:
    # Contar pedidos por estado (top 15)
    pedidos_por_estado = (
        df['customer_state']
        .value_counts()
        .head(15)
    )
    
    # Criar figura e eixo
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Gráfico de barras horizontal (melhor para muitas categorias)
    sns.barplot(
        y=pedidos_por_estado.index,
        x=pedidos_por_estado.values,
        palette="viridis",
        ax=ax
    )
    
    # Configurações de estilo
    ax.set_title(
        "São Paulo domina as vendas online no Brasil (2016-2018)",
        fontsize=14, fontweight='bold', pad=20
    )
    ax.set_xlabel("Número de Pedidos", fontsize=12)
    ax.set_ylabel("Estado", fontsize=12)
    
    # Adicionar valores nas barras
    for i, v in enumerate(pedidos_por_estado.values):
        ax.text(v + 500, i, f'{v:,}', va='center', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Insight: {pedidos_por_estado.index[0]} concentra {pedidos_por_estado.values[0]:,} pedidos,")
    print(f"   representando {(pedidos_por_estado.values[0] / df.shape[0] * 100):.1f}% do total.")
````

### 9.2 DISTRIBUIÇÃO — Boxplot: Valores por Status

**Objetivo:** Entender como o valor médio dos pedidos varia conforme seu status final.

**Por que esse gráfico?** Um boxplot mostra mediana, quartis e outliers em um único gráfico compacto — ideal para comparar distribuições entre grupos.

**Pergunta de negócio:** Pedidos cancelados tendem a ter valores diferentes dos entregues?

````python
if df is not None and 'order_value' in df.columns:
    # Filtrar para os status mais comuns e remover nulos
    top_status = df['order_status'].value_counts().head(5).index.tolist()
    df_box = df[df['order_status'].isin(top_status)].dropna(subset=['order_value'])
    
    # Remover outliers extremos para melhor visualização (percentil 99)
    limite = df_box['order_value'].quantile(0.99)
    df_box = df_box[df_box['order_value'] <= limite]
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    sns.boxplot(
        data=df_box,
        x='order_status',
        y='order_value',
        palette='Set2',
        ax=ax
    )
    
    ax.set_title(
        "Distribuição de Valores de Pedidos por Status",
        fontsize=14, fontweight='bold', pad=20
    )
    ax.set_xlabel("Status do Pedido", fontsize=12)
    ax.set_ylabel("Valor do Pedido (R$)", fontsize=12)
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    # Estatísticas por status
    print("\n📊 Estatísticas por Status:")
    print(df_box.groupby('order_status')['order_value'].describe().round(2))
````

### 9.3 DISTRIBUIÇÃO — Violin Plot (ementa)

**Objetivo:** Visualizar a **forma completa da distribuição** dos valores por status.

**Por que esse gráfico?** O violin plot combina boxplot com densidade (KDE). A "largura" do violino mostra onde os dados se concentram — é mais informativo que um boxplot puro quando queremos ver a **forma** da distribuição, não apenas os quartis.

**Quando usar boxplot vs violin plot?**
- **Boxplot**: quando o foco é em mediana, quartis e outliers
- **Violin plot**: quando o foco é na **forma** (bimodal, assimétrica, etc.)

**Pergunta de negócio:** Como se comporta a distribuição de notas de avaliação para pedidos entregues vs atrasados?

````python
if df is not None and 'review_score' in df.columns and 'prazo_entrega_dias' in df.columns:
    # Criar categoria de prazo
    df_vio = df.dropna(subset=['review_score', 'prazo_entrega_dias']).copy()
    
    # Categorizar prazo de entrega
    def categorizar_prazo(dias):
        if dias <= 7:
            return 'Rápido (≤7d)'
        elif dias <= 15:
            return 'Normal (8-15d)'
        elif dias <= 30:
            return 'Lento (16-30d)'
        else:
            return 'Muito lento (>30d)'
    
    df_vio['categoria_prazo'] = df_vio['prazo_entrega_dias'].apply(categorizar_prazo)
    ordem = ['Rápido (≤7d)', 'Normal (8-15d)', 'Lento (16-30d)', 'Muito lento (>30d)']
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    sns.violinplot(
        data=df_vio,
        x='categoria_prazo',
        y='review_score',
        order=ordem,
        palette='RdYlGn_r',  # vermelho->amarelo->verde invertido
        ax=ax
    )
    
    ax.set_title(
        "Forma da distribuição de avaliações conforme prazo de entrega",
        fontsize=14, fontweight='bold', pad=20
    )
    ax.set_xlabel("Categoria de Prazo de Entrega", fontsize=12)
    ax.set_ylabel("Nota da Avaliação (1-5)", fontsize=12)
    ax.set_ylim(0.5, 5.5)
    
    plt.tight_layout()
    plt.show()
    
    print("\n📊 Insight: Observe como entregas rápidas concentram notas 5,")
    print("   enquanto entregas muito lentas têm distribuição 'empurrada' para notas baixas.")
    print("\nMédias por categoria:")
    print(df_vio.groupby('categoria_prazo')['review_score'].agg(['mean', 'count']).round(2))
````

### 9.4 DISTRIBUIÇÃO — Histograma com Densidade

**Objetivo:** Entender como os valores de pedidos se distribuem.

**Por que esse gráfico?** Um histograma com curva de densidade mostra tanto a frequência quanto a forma da distribuição, revelando se há picos, assimetrias ou múltiplas modas.

**Pergunta de negócio:** Qual é o padrão de gasto dos clientes? Há concentração em faixas específicas?

````python
if df is not None and 'order_value' in df.columns:
    df_hist = df.dropna(subset=['order_value'])
    
    # Limitar outliers extremos (percentil 99) para melhor visualização
    limite = df_hist['order_value'].quantile(0.99)
    df_hist = df_hist[df_hist['order_value'] <= limite]
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    sns.histplot(
        data=df_hist, x='order_value',
        kde=True, stat='density', bins=50,
        color='skyblue', ax=ax
    )
    
    media = df_hist['order_value'].mean()
    mediana = df_hist['order_value'].median()
    
    ax.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Média: R${media:.2f}')
    ax.axvline(mediana, color='green', linestyle='--', linewidth=2, label=f'Mediana: R${mediana:.2f}')
    
    ax.set_title(
        "Distribuição dos Valores de Pedidos",
        fontsize=14, fontweight='bold', pad=20
    )
    ax.set_xlabel("Valor do Pedido (R$)", fontsize=12)
    ax.set_ylabel("Densidade", fontsize=12)
    ax.legend(fontsize=10)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Insights:")
    print(f"  • Média: R${media:.2f}")
    print(f"  • Mediana: R${mediana:.2f}")
    print(f"  • Assimetria (skewness): {df_hist['order_value'].skew():.2f}")
    print(f"  • Distribuição é assimétrica à direita (cauda longa de valores altos)")
````

### 9.5 RELAÇÃO — Série Temporal de Vendas

**Objetivo:** Visualizar a evolução das vendas ao longo do tempo.

**Por que esse gráfico?** Um gráfico de linha mostra tendências, sazonalidade e quebras estruturais na série temporal.

**Pergunta de negócio:** Como evoluem as vendas? Há padrões sazonais?

````python
if df is not None and 'order_year_month' in df.columns:
    vendas_por_mes = (
        df.dropna(subset=['order_year_month', 'order_value'])
        .groupby('order_year_month')
        .agg(num_pedidos=('order_id', 'count'), receita=('order_value', 'sum'))
    )
    vendas_por_mes.index = vendas_por_mes.index.astype(str)
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 9))
    
    # Número de pedidos
    ax1.plot(range(len(vendas_por_mes)), vendas_por_mes['num_pedidos'].values,
             marker='o', linewidth=2.5, markersize=6, color='#2E86AB')
    ax1.fill_between(range(len(vendas_por_mes)), vendas_por_mes['num_pedidos'].values,
                      alpha=0.3, color='#2E86AB')
    ax1.set_title("Volume de Pedidos ao Longo do Tempo", fontsize=12, fontweight='bold')
    ax1.set_ylabel("Número de Pedidos", fontsize=11)
    ax1.grid(True, alpha=0.3)
    ax1.set_xticks(range(0, len(vendas_por_mes), 3))
    ax1.set_xticklabels(vendas_por_mes.index[::3], rotation=45)
    
    # Receita total
    ax2.plot(range(len(vendas_por_mes)), vendas_por_mes['receita'].values,
             marker='s', linewidth=2.5, markersize=6, color='#A23B72')
    ax2.fill_between(range(len(vendas_por_mes)), vendas_por_mes['receita'].values,
                      alpha=0.3, color='#A23B72')
    ax2.set_title("Receita Total ao Longo do Tempo", fontsize=12, fontweight='bold')
    ax2.set_xlabel("Período", fontsize=11)
    ax2.set_ylabel("Receita (R$)", fontsize=11)
    ax2.grid(True, alpha=0.3)
    ax2.set_xticks(range(0, len(vendas_por_mes), 3))
    ax2.set_xticklabels(vendas_por_mes.index[::3], rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Padrões temporais:")
    print(f"  • Mês com mais pedidos: {vendas_por_mes['num_pedidos'].idxmax()} ({vendas_por_mes['num_pedidos'].max():,})")
    print(f"  • Mês com maior receita: {vendas_por_mes['receita'].idxmax()} (R${vendas_por_mes['receita'].max():,.2f})")
````

### 9.6 RELAÇÃO — Regressão Linear (ementa)

**Objetivo:** Investigar se existe **relação linear** entre duas variáveis.

**Por que esse gráfico?** A função `sns.regplot()` ajusta automaticamente uma **linha de regressão linear** aos dados e mostra o **intervalo de confiança** (área sombreada). É uma forma poderosa de visualizar tendências e avaliar visualmente a força da correlação.

**Como ler:**
- **Inclinação da linha**: positiva = quando X aumenta, Y aumenta; negativa = inverso
- **Dispersão dos pontos**: pouca dispersão = relação forte; muita dispersão = relação fraca
- **Área sombreada**: quanto menor, mais confiável o ajuste

**Pergunta de negócio:** Clientes que gastam mais estão mais satisfeitos?

````python
if df is not None and 'review_score' in df.columns and 'order_value' in df.columns:
    df_reg = df.dropna(subset=['order_value', 'review_score'])
    
    # Limitar outliers
    limite = df_reg['order_value'].quantile(0.99)
    df_reg = df_reg[df_reg['order_value'] <= limite]
    
    # Amostragem para não sobrecarregar o gráfico
    df_sample = df_reg.sample(min(5000, len(df_reg)), random_state=42)
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    sns.regplot(
        data=df_sample,
        x='order_value',
        y='review_score',
        scatter_kws={'alpha': 0.3, 's': 20, 'color': '#457B9D'},
        line_kws={'color': '#E63946', 'linewidth': 2.5},
        ax=ax
    )
    
    ax.set_title(
        "Regressão Linear: Valor do Pedido vs Avaliação",
        fontsize=14, fontweight='bold', pad=20
    )
    ax.set_xlabel("Valor do Pedido (R$)", fontsize=12)
    ax.set_ylabel("Avaliação do Cliente (1-5)", fontsize=12)
    
    # Calcular correlação
    correlacao = df_reg['order_value'].corr(df_reg['review_score'])
    ax.text(0.05, 0.95, f'Correlação de Pearson: {correlacao:.3f}',
            transform=ax.transAxes, fontsize=11, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.7))
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Análise de correlação:")
    print(f"  • Correlação de Pearson: {correlacao:.4f}")
    if abs(correlacao) < 0.3:
        print(f"  • Interpretação: Correlação FRACA entre valor e avaliação")
        print(f"    → O valor do pedido NÃO é um bom preditor da satisfação")
    elif abs(correlacao) < 0.7:
        print(f"  • Interpretação: Correlação MODERADA")
    else:
        print(f"  • Interpretação: Correlação FORTE")
````

### 9.7 RELAÇÃO — Heatmap de Correlação

**Objetivo:** Visualizar correlações entre múltiplas variáveis numéricas em uma única visão.

**Por que esse gráfico?** O heatmap mostra, em uma matriz colorida, como todas as variáveis numéricas se correlacionam entre si. É perfeito para detectar rapidamente quais variáveis estão relacionadas.

**Como ler:**
- Cores próximas de **+1** (vermelho escuro): correlação **positiva forte**
- Cores próximas de **-1** (azul escuro): correlação **negativa forte**
- Cores próximas de **0** (branco): **sem correlação**

**Pergunta de negócio:** Qual é a estrutura de correlação entre variáveis do negócio?

````python
if df is not None:
    # Selecionar colunas numéricas relevantes
    colunas_num = ['order_value', 'num_items', 'review_score', 'prazo_entrega_dias', 'order_year', 'order_month']
    colunas_disponiveis = [c for c in colunas_num if c in df.columns]
    
    if len(colunas_disponiveis) > 1:
        corr_matrix = df[colunas_disponiveis].corr()
        
        fig, ax = plt.subplots(figsize=(10, 8))
        
        sns.heatmap(
            corr_matrix,
            annot=True, fmt='.2f',
            cmap='coolwarm',  # paleta divergente
            center=0,
            square=True,
            linewidths=1,
            cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1,
            ax=ax
        )
        
        ax.set_title(
            "Matriz de Correlação entre Variáveis Numéricas",
            fontsize=14, fontweight='bold', pad=20
        )
        
        plt.tight_layout()
        plt.show()
        
        print("\n📊 Destaques:")
        print("  • Observe quais variáveis têm correlação mais forte (positiva ou negativa)")
        print("  • Correlações próximas de zero = variáveis independentes")
````

### 9.8 COMPOSIÇÃO — Pizza Chart: Status dos Pedidos

**Objetivo:** Mostrar a proporção de pedidos em cada status.

**Por que esse gráfico?** Um gráfico de pizza é útil quando há poucas categorias (até 5) e queremos destacar proporções visuais.

**Pergunta de negócio:** Qual é a taxa de sucesso? Quantos pedidos são cancelados?

````python
if df is not None and 'order_status' in df.columns:
    status_counts = df['order_status'].value_counts()
    
    fig, ax = plt.subplots(figsize=(10, 8))
    
    cores = sns.color_palette("husl", len(status_counts))
    
    wedges, texts, autotexts = ax.pie(
        status_counts.values,
        labels=status_counts.index,
        autopct='%1.1f%%',
        colors=cores,
        startangle=90,
        textprops={'fontsize': 11}
    )
    
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontweight('bold')
    
    ax.set_title(
        "Distribuição de Pedidos por Status (2016-2018)",
        fontsize=14, fontweight='bold', pad=20
    )
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Resumo por Status:")
    for status, count in status_counts.items():
        pct = (count / status_counts.sum()) * 100
        print(f"  • {status.upper()}: {count:,} pedidos ({pct:.1f}%)")
````

### 9.9 COMPOSIÇÃO — Barras Empilhadas: Composição por Estado

**Objetivo:** Mostrar como o status dos pedidos se distribui dentro de cada estado.

**Por que esse gráfico?** Uma barra empilhada permite comparar composição entre grupos mantendo visibilidade das partes.

**Pergunta de negócio:** Alguns estados têm maior taxa de cancelamento?

````python
if df is not None:
    top_estados = df['customer_state'].value_counts().head(10).index
    
    crosstab = pd.crosstab(
        df[df['customer_state'].isin(top_estados)]['customer_state'],
        df[df['customer_state'].isin(top_estados)]['order_status'],
        normalize='index'
    ) * 100
    
    fig, ax = plt.subplots(figsize=(14, 7))
    
    crosstab.plot(
        kind='barh', stacked=True, ax=ax,
        color=sns.color_palette("Set2", len(crosstab.columns))
    )
    
    ax.set_title(
        "Composição do Status dos Pedidos por Estado (%)",
        fontsize=14, fontweight='bold', pad=20
    )
    ax.set_xlabel("Percentual (%)", fontsize=12)
    ax.set_ylabel("Estado", fontsize=12)
    ax.legend(title="Status", bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📊 Taxa de Entrega por Estado (Top 10):")
    print(crosstab.round(2))
````

## 10. Comunicação Visual e Ética

Visualização de dados também é uma questão de **responsabilidade**. Um gráfico pode informar corretamente, mas também pode induzir interpretações equivocadas quando é construído de forma ruim.

### 10.1 Como evitar gráficos enganosos

#### ❌ NÃO faça:
- **Manipular escalas** para exagerar diferenças mínimas
- **Omitir o zero** em gráficos de barras (exceto em heatmaps)
- **Usar cores** sem propósito (cores devem categorizar ou destacar)
- **Truncar eixos** para fazer um padrão parecer mais forte
- **Omitir contexto relevante** (por exemplo, esconder valores negativos)
- **Usar gráfico de pizza com mais de 5 categorias** (fica ilegível)
- **Deixar de documentar filtros ou agregações**

#### ✅ FAÇA:
- Sempre começar os eixos em zero (ou deixar bem claro por que não)
- Incluir título descritivo e fonte dos dados
- Usar cores consistentemente
- Fornecer rótulos claros
- Indicar quando há dados filtrados ou faltando
- Escolher o tipo de gráfico apropriado para a pergunta
- Testar a visualização com usuários não técnicos

### 10.2 Exemplo: o perigo de manipular escalas

````python
# Exemplo: gráfico manipulado vs. honesto
dados_exemplo = pd.DataFrame({
    'Trimestre': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Vendas': [100, 102, 105, 103]
})

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico ENGANOSO (escala manipulada)
ax1.bar(dados_exemplo['Trimestre'], dados_exemplo['Vendas'], color='red', alpha=0.7)
ax1.set_ylim(99, 106)  # Truncando a escala!
ax1.set_title("❌ ENGANOSO: Escala manipulada\n(parece crescimento explosivo)",
              fontsize=12, fontweight='bold', color='red')
ax1.set_ylabel("Vendas")

# Gráfico HONESTO (escala apropriada)
ax2.bar(dados_exemplo['Trimestre'], dados_exemplo['Vendas'], color='green', alpha=0.7)
ax2.set_ylim(0, 110)  # Começando do zero
ax2.set_title("✅ HONESTO: Escala apropriada\n(crescimento modesto visível)",
              fontsize=12, fontweight='bold', color='green')
ax2.set_ylabel("Vendas")

plt.suptitle("O perigo de manipular escalas em visualizações",
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("\n⚠️  Lição: A mesma informação pode ser apresentada de formas radicalmente diferentes.")
print("   Como visualizador, você tem a responsabilidade de ser honesto com os dados.")
````

## 11. Checklist: Como fazer uma visualização de qualidade

Antes de apresentar um gráfico, pergunte a si mesmo:

### Clareza e Propósito
- [ ] O gráfico responde a uma pergunta clara?
- [ ] O título é específico e informativo (não apenas "Vendas por Estado")?
- [ ] O público-alvo consegue entender sem ajuda adicional?

### Escolha do Gráfico
- [ ] É o tipo de gráfico mais apropriado para essa pergunta?
- [ ] Estou comparando, mostrando distribuição, relação ou composição?
- [ ] Evitei gráficos "bonitos mas inúteis"?

### Design Visual
- [ ] As cores têm propósito (não são apenas decoração)?
- [ ] A paleta escolhida é adequada (qualitativa, sequencial ou divergente)?
- [ ] O gráfico funciona em preto e branco (acessibilidade)?
- [ ] Os rótulos são legíveis (tamanho, rotação, sobreposição)?
- [ ] A paleta de cores respeita usuários com daltonismo?

### Dados e Honestidade
- [ ] A escala começa em zero (quando apropriado)?
- [ ] Não há manipulação intencional para exagerar efeitos?
- [ ] A fonte dos dados está clara?
- [ ] Filtros ou agregações estão documentados?
- [ ] Valores faltantes ou nulos foram tratados transparentemente?

### Contexto
- [ ] O gráfico está inserido em uma narrativa maior?
- [ ] Há anotações para destacar pontos importantes?
- [ ] A legenda é necessária ou redundante?
